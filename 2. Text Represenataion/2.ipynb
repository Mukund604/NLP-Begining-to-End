>## BAG OF WORDS
>>We make a vocabulary of all the top features present in all the sentences, after that unlike one hot encoding, we have a fixed bag of vocabulary, then for each sentence we ask the how many times this word occurs in the sentence. We count the number of occurances from the sentences, to make vectors out of the textual data. For eg : please write a comment and please share. Here the ferquency of please is 2 and rest have the frequency 1. This eliminates the issue of OOV (Out of Vocab) Error as we are asking the question of how may times this word in the vocabulay has occured.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

Bow = cv.fit_transform(testing_df['review'])
```

>## Bag Of Words N-grams, bi-grams, tri-grams
>> Works like Bag of words only but the difference being, when we make the vocabulary, we dont use single words to form the vocabulary, we use bi(two), tri(three) or n (any number of words), to form the vocaublary. These are called bi, tri and n - Grams

```python
bi_gram_Cv = CountVectorizer(ngram_range=(3,3))
Bi_gram_Bow = bi_gram_Cv.fit_transform(df['review'])
bi_gram_Cv.vocabulary_
```

>#  TF-IDF (Term Frequency - Inverse Document Frequency)
>> Assigns different values to different words, works on the logic, if a word is of very high frequency in that particular document, but is very rare in the full corpus, then that word holds a lot of significance for that particular document. so we give that word more value as compared to other words.
>>### 1. Term Frequency = (Number of occurances of term t in a document / Total number of words in the document)
>>### 2. Inverse Document Frequency = log(Total number of documents in the corpus / Number of documents with the term t in them)
>>We add a 1 to the prodcut of TF and IDF so that the term which occurs in every document isn't ignored. This is because IDF of the term which occurs in every document is 0, so 0 product with the TF makes it 0 and thus ignores the word completely
